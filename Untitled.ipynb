```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
mono_pred_disparities = np.load("./predict/reference/pickle/disparities_pp.npy")
falnet_pred_disparities = np.load("./predict/kitti_sample3/pickle/predictions.pickle", allow_pickle=True)
```

```python
print(mono_pred_disparities.shape)
print(falnet_pred_disparities[0].shape)
```

```python
def visualize_disparity(disparity):
    disparity = 256 * np.clip(disparity / (np.percentile(disparity, 99) + 1e-9), 0, 1)
    plt.imshow(np.rint(disparity).astype(np.int32),
               cmap="inferno",
               vmin=0,
               vmax=256,
    )
    plt.show()
```

```python
def rescale(value, rmin, rmax, tmin, tmax):
    return ((value - rmin) / (rmax - rmin)) * (tmax - tmin) + tmin
```

```python
type(falnet_pred_disparities[0])
```

```python
image = falnet_pred_disparities[0]
visualize_disparity(image)

hshift = 60
vshift = 600
visualize_disparity(image[0+hshift:256+hshift, 0+vshift:512+vshift])

image = cv2.resize(image, dsize=(512, 256), interpolation=cv2.INTER_AREA)
visualize_disparity(image)
```

```python
visualize_disparity(mono_pred_disparities[0])
```

```python
resized = [cv2.resize(image, dsize=(512, 256), interpolation=cv2.INTER_AREA) for image in falnet_pred_disparities]
```

```python
new_falnet = np.empty_like(mono_pred_disparities)
for idx, image in enumerate(resized):
    new_falnet[idx] = image
```

```python
new_falnet.shape
visualize_disparity(new_falnet[0])
```

```python
np.save("./resized_falnet_predictions.npy", new_falnet)
```
